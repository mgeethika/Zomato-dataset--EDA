```python
df_country = pd.read_excel('Country-Code.xlsx') #to merge the counntry code excel sheet, reading excel file
```

```python
df_country.head() #first 5 records
```

```python
new_df=pd.merge(df, df_country, on ='Country Code', how='left') #merging the sheets
```

```python
new_df.columns
```

```python
 new_df.Country.value_counts() #countries and the count of transactions/usage
```

```python
country_names = new_df.Country.value_counts().index #splitting index as names
```

```python
country_val = new_df.Country.value_counts().values #splitting values as val
```

```python
plt.pie(country_val[:3], labels=country_names[:3], autopct='%1.2f%%') 
#displaying only top 3 with percentage uptp 2 decimals
```

```python
new_df.columns
```

```python
new_df.groupby(['Aggregate rating', 'Rating color','Rating text' ]).size() #to count how many ratings
```

```python
new_df.groupby(['Aggregate rating', 'Rating color','Rating text' ]).size().reset_index().rename(columns = {0:'count'})
#renaming the column as count
```

```python
ratings= new_df.groupby(['Aggregate rating', 'Rating color','Rating text' ]).size().reset_index().rename(columns = {0:'count'})
#assigning a new name to df
```

```python
ratings
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)  #adjust size
sns.barplot(data=ratings, x ="Aggregate rating" ,hue = 'Rating color', y="count", palette=['Blue', 'red', 'orange','yellow','green','green'])
#displaying the rating colours to match the barplot colors
```

```python
sns.countplot(data=ratings, x="Rating color",palette=['Blue', 'red', 'orange','yellow','green','green']) #frequency
```
